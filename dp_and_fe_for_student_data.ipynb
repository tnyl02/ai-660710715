```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import numpy as np
```

```python
df = pd.read_csv("student.csv")
df
```

```python
new_data = df.copy()
```

```python
new_data['ชื่อ'] = new_data['ชื่อ'].fillna('ไม่ระบุ')
new_data
```

```python
def standardize_gender(เพศ):
    if เพศ in ['หญิง', 'Female', 'F', 'ญ']:
        return 'หญิง'
    elif เพศ in ['ชาย', 'ช', 'Male', 'M']:
        return 'ชาย'
    else:
        return เพศ

new_data['เพศ'] = new_data['เพศ'].apply(standardize_gender)
new_data
```

```python
def standardize_faculty(คณะ):
    if คณะ in ['วิทยาศาสตร์', 'วิทย์', 'คณะวิทยาศาสตร์', 'Science', 'Sci']:
        return 'วิทยาศาสตร์'
    elif คณะ in ['วิศวะ', 'Engineering', 'วิศวกรรม', 'วิศวฯ', 'วิศวกรรมศาสตร์']:
        return 'วิศวกรรมศาสตร์'
    elif คณะ in ['ศิลปศาสตร์', 'Arts', 'ศิลป์', 'Liberal Arts']:
        return 'ศิลปศาสตร์'
    else:
        return คณะ

new_data['คณะ'] = new_data['คณะ'].apply(standardize_faculty)
new_data
```

```python
mean_gpa = new_data['GPA'].mean()
new_data.loc[new_data['GPA'] < 0, 'GPA'] = np.nan
new_data.loc[new_data['GPA'] > 4, 'GPA'] = 4.0
new_data['GPA'].fillna(mean_gpa, inplace=True)
new_data
```

```python
mean_19_23 = new_data.loc[(new_data['อายุ'] >= 19) & (new_data['อายุ'] <= 23), 'อายุ'].mean()
condition = (new_data['อายุ'] > 23) | (new_data['อายุ'].isna()) | (new_data['อายุ'] == "")
new_data.loc[condition, 'อายุ'] = mean_19_23
new_data
```

```python
maxhours = new_data['ชั่วโมงเรียน_ต่อสัปดาห์'].max()
minhours = new_data['ชั่วโมงเรียน_ต่อสัปดาห์'].min()
print(f"ชั่วโมงเรียน_ต่อสัปดาห์: สูงสุด = {maxhours}, ต่ำสุด = {minhours}")
meanHR = new_data['ชั่วโมงเรียน_ต่อสัปดาห์'].mean()
```

```python
new_data['ชั่วโมงเรียน_ต่อสัปดาห์'].fillna(meanHR, inplace=True)
new_data
```

```python
new_data['BMI'] = new_data['น้ำหนัก_kg'] / (new_data['ส่วนสูง_cm']/100)**2
new_data['สุขภาพ'] = pd.cut(new_data['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])

print(new_data[['ชื่อ', 'BMI', 'สุขภาพ']])
```

```python
Reg = new_data.copy()
```

```python
Reg['วันที่ลงทะเบียน'] = pd.to_datetime(Reg['วันที่ลงทะเบียน'])
```

```python
count_reg = Reg.groupby(Reg['วันที่ลงทะเบียน'].dt.date).size().reset_index(name='จำนวนนศ')

print("แต่ละวันที่นักศึกษาได้ลงทะเบียนเรียน")
print(count_reg)
```
